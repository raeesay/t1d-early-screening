```python
# Load libraries and data
import pandas as pd

data = pd.read_excel('dataset1.xlsx')
# 10374 rows, 29 columns
print(data.shape)
data.head()
```

```python
# Check for single screening negatives
# uid only once in dataset with final_report_screening negative

# Count how often each uid appears
uid_counts = data['uid'].value_counts()

# Get list of uid that appear exactly once
uids_once = uid_counts[uid_counts == 1].index
# We have 3280 uid that show up only once
# print(uids_once)

# Double check their final report
df_once = data[data['uid'].isin(uids_once)]
print(df_once['all_reports_screening'].value_counts())
print(df_once['final_report_screening'].value_counts())
```

```python
# Get all single screening negative reports
df_single_screening_neg = df_once[df_once['all_reports_screening'] == 'negative']
# df_single_screening_neg.to_excel("dataset1_single_screening_neg.xlsx", index=False)
```

```python
# Check how many negative reports have t1d_date as last_contact_date_type
print(df_single_screening_neg[df_single_screening_neg['last_contact_date_type'] == 't1d_date'].shape)
# There are 11 instances where despite negative single screening the last_contact_date_type is t1d_date
# These are most likely mistakes

# Check how many are hemolyzed -> 1132
print(df_single_screening_neg[df_single_screening_neg['hemolyzed'] == 1].shape)
# Check how many samples were reevaluated (should be none) -> 0
print(df_single_screening_neg[df_single_screening_neg['reevaluated'] == 1].shape)

```
