```python
# Add ELISA gating according to your established rules
def compute_elisa_pos(df):
    df = df.copy()
    df["sample_date"] = pd.to_datetime(df.get("sample_date"))  # only if available

    # If sample_date not present in clean_df, fall back to a global rule.
    # But your earlier code did have sample_date.

    # If sample_date exists in your clean_df:
    if "sample_date" in df.columns:
        df["elisa_pos"] = np.select(
            [
                df["sample_date"] <= pd.Timestamp("2022-01-31"), 
                df["sample_date"].between("2022-02-01", "2024-12-31"),
                df["sample_date"] > pd.Timestamp("2024-12-31")
            ],
            [
                df["elisa"] >= 25,
                df["elisa"] >= 40,
                df["elisa"] >= 35
            ],
            default=False
        )
    else:
        # If sample_date not in clean_df: use a fallback
        df["elisa_pos"] = df["elisa"] >= 40

    return df

clean_df = compute_elisa_pos(clean_df)
clean_df.head()
```

### 1. Bayesian Regression

```python
# ------------------------------------------------------------
# Define which rows contain RIA data (replaces the ELISA filter)
# ------------------------------------------------------------
RIA_SOURCES = {"screening", "confirmation", "follow_up", "antibody_follow_up"}

def fit_bayesian_mixture_cutoff(
    df,
    biomarker_col,
    label_col="label_early_stage",
    prevalence=None,
    grid_points=200,
    draws=2000,
    tune=2000,
    chains=4,
    target_accept=0.9,
    random_seed=42,
):
    """
    Two–group lognormal mixture model for RIA antibody titers.
    """

    # --------------------------------------------------------
    # 1. Prepare data
    # --------------------------------------------------------
    sub = df.copy()

    # Use only rows where RIA is actually measured
    sub = sub[sub["source"].isin(RIA_SOURCES)]

    # Require biomarker + label
    sub = sub.dropna(subset=[biomarker_col, label_col]).copy()

    x_raw = sub[biomarker_col].astype(float).values
    y = sub[label_col].astype(int).values

    # Must have positive values for lognormal
    positive_mask = x_raw > 0
    if positive_mask.sum() == 0:
        raise ValueError(f"No positive values for {biomarker_col} in RIA data.")

    x_raw = x_raw[positive_mask]
    y = y[positive_mask]

    # Avoid log(0)
    eps = np.percentile(x_raw, 0.1)
    x_clip = np.clip(x_raw, eps, None)
    log_x = np.log(x_clip)

    # Prevalence = proportion of true early-stage in RIA subset
    if prevalence is None:
        prevalence = y.mean()

    print(f"\n=== RIA-MIX MODEL {biomarker_col} ===")
    print(f"N = {len(x_raw)}, positive rate = {y.mean():.3f}, prevalence = {prevalence:.3f}")

    # --------------------------------------------------------
    # 2. PyMC model
    # --------------------------------------------------------
    with pm.Model() as model:

        # Negative distribution (lognormal)
        mu_neg = pm.Normal("mu_neg", 0, 2)
        sigma_neg = pm.HalfNormal("sigma_neg", 2)

        # Positive mixture (two components)
        w = pm.Dirichlet("w", a=np.array([1.0, 1.0]))

        mu_pos1 = pm.Normal("mu_pos1", 1.0, 2.0)
        sigma_pos1 = pm.HalfNormal("sigma_pos1", 2.0)

        mu_pos2 = pm.Normal("mu_pos2", 3.0, 2.0)
        sigma_pos2 = pm.HalfNormal("sigma_pos2", 2.0)

        # Split observations
        log_x_neg = log_x[y == 0]
        log_x_pos = log_x[y == 1]

        pm.Normal("neg_like", mu=mu_neg, sigma=sigma_neg, observed=log_x_neg)

        pm.NormalMixture(
            "pos_like",
            w=w,
            mu=pm.math.stack([mu_pos1, mu_pos2]),
            sigma=pm.math.stack([sigma_pos1, sigma_pos2]),
            observed=log_x_pos,
        )

        trace = pm.sample(
            draws=draws,
            tune=tune,
            chains=chains,
            target_accept=target_accept,
            random_seed=random_seed,
        )

    print(az.summary(trace, var_names=["mu_neg","sigma_neg","mu_pos1","sigma_pos1","mu_pos2","sigma_pos2","w"]))

    # --------------------------------------------------------
    # 3. Posterior PPV curve
    # --------------------------------------------------------
    grid_raw = np.linspace(x_raw.min(), x_raw.max(), grid_points)
    grid_clip = np.clip(grid_raw, eps, None)
    grid_log = np.log(grid_clip)

    posterior = trace.posterior
    mu_neg_s = posterior["mu_neg"].stack(sample=("chain","draw")).values
    sigma_neg_s = posterior["sigma_neg"].stack(sample=("chain","draw")).values

    mu_pos1_s = posterior["mu_pos1"].stack(sample=("chain","draw")).values
    sigma_pos1_s = posterior["sigma_pos1"].stack(sample=("chain","draw")).values
    mu_pos2_s = posterior["mu_pos2"].stack(sample=("chain","draw")).values
    sigma_pos2_s = posterior["sigma_pos2"].stack(sample=("chain","draw")).values

    w_s = posterior["w"].stack(sample=("chain","draw")).values  # shape (2, samples)

    dens_neg = []
    dens_pos = []

    for z, xval in zip(grid_log, grid_clip):
        d_neg = norm.pdf(z, mu_neg_s, sigma_neg_s) / xval
        d_pos1 = norm.pdf(z, mu_pos1_s, sigma_pos1_s) / xval
        d_pos2 = norm.pdf(z, mu_pos2_s, sigma_pos2_s) / xval

        d_pos = w_s[0]*d_pos1 + w_s[1]*d_pos2

        dens_neg.append(d_neg)
        dens_pos.append(d_pos)

    dens_neg = np.array(dens_neg)
    dens_pos = np.array(dens_pos)

    pi = prevalence
    ppv_samples = (dens_pos * pi) / (dens_pos * pi + dens_neg * (1-pi))

    ppv_mean = ppv_samples.mean(axis=1)
    ppv_low, ppv_high = np.percentile(ppv_samples, [5, 95], axis=1)

    # --------------------------------------------------------
    # 4. Extract PPV cutoffs
    # --------------------------------------------------------
    def cutoff(ppv_target):
        idx = np.where(ppv_mean >= ppv_target)[0]
        return grid_raw[idx[0]] if len(idx)>0 else np.nan

    cutoffs = {
        "ppv_0.01": cutoff(0.01),
        "ppv_0.10": cutoff(0.10),
        "ppv_0.50": cutoff(0.50),
        "ppv_0.995": cutoff(0.995),
    }

    return {
        "biomarker": biomarker_col,
        "grid_raw": grid_raw,
        "ppv_mean": ppv_mean,
        "ppv_low": ppv_low,
        "ppv_high": ppv_high,
        "cutoffs": cutoffs,
        "trace": trace,
        "model": model,
    }
```
